# Packages

```python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import seaborn           as sns
import os
```

# Paths

```python
outputs_path = '../../LargeFilesResults/SEDFitting/'
# new_dir = os.mkdir(os.path.join(outputs_path, 'Analyses_mdegrees'))
```

# Reading the outputs

```python
out_m150 = pd.read_csv(os.path.join(outputs_path, 'Batch150/outputs_ppxf_150_4350.csv'))
out_m200 = pd.read_csv(os.path.join(outputs_path, 'Batch200/outputs_ppxf_200_4350.csv'))
out_m230 = pd.read_csv(os.path.join(outputs_path, 'Batch230/outputs_ppxf_230_4350.csv'))
```

# Merging them into a single dataframe

```python
out_m230.columns = ['CATAID', 'Z', '<logAge>_m230', '<[M/H]>_m230', 'MASS2LIGHT_R_m230', 'S2N_runz', 'CHI2_PPXF_m230']
```

```python
mega_dataframe = out_m150.merge(out_m200, how='inner', on=['CATAID', 'Z'], suffixes=('_m150', '_m200'),
                                validate='one_to_one').merge(out_m230, how='inner', on=['CATAID', 'Z'], 
                                                             validate='one_to_one')
```

```python
mega_dataframe.to_csv(os.path.join(outputs_path, 'summary_batches.csv'))
```

# Analyses

```python
fontsize = 15
```

```python
plt.plot(out_m230['S2N_runz'], out_m230['CHI2_PPXF_m230'], 'o')
plt.xlabel('S/N (runz)', fontsize=fontsize)
plt.ylabel(r"pPXF $\chi^2$", fontsize=fontsize)
plt.tick_params(axis='both', labelsize=fontsize)
plt.savefig(os.path.join(outputs_path, 'Analyses_mdegrees/chi_snI.png'))
plt.savefig(os.path.join(outputs_path, 'Analyses_mdegrees/chi_snI.pdf'))
plt.show()
```

```python
top_s2n_runz   = mega_dataframe.loc[mega_dataframe['S2N_runz']>10]
worst_s2n_runz = mega_dataframe.loc[mega_dataframe['S2N_runz']<10]
```

```python
top_s2n_runz.values.shape
```

```python
worst_s2n_runz.values.shape
```

```python
plt.subplots(figsize=(8, 5))
plt.plot(top_s2n_runz['S2N_runz'], top_s2n_runz['CHI2_PPXF_m230'], 'o')
plt.plot(worst_s2n_runz['S2N_runz'], worst_s2n_runz['CHI2_PPXF_m230'], 'o')
plt.xlabel('S/N (runz)', fontsize=fontsize)
plt.ylabel(r"pPXF $\chi^2$", fontsize=fontsize)
plt.tick_params(axis='both', labelsize=fontsize)
plt.savefig(os.path.join(outputs_path, 'Analyses_mdegrees/ch_snII.png'))
plt.savefig(os.path.join(outputs_path, 'Analyses_mdegrees/ch_snII.pdf'))
plt.tight_layout()
plt.show()
```

```python
figsize = (10,5)
pal_top = ['#edf8b1','#7fcdbb','#2c7fb8']
pal_wst = ['#fee8c8','#fdbb84','#e34a33']
alpha = 0.7
xlim = [7.5, 10.5]
ylim = [-1.2, 0.0]
```

```python
plt.subplots(figsize=figsize)

ax1 = plt.subplot(1, 3, 1)
plt.plot(top_s2n_runz['<logAge>_m150'], top_s2n_runz['<[M/H]>_m150'], 'o', color=pal_top[2], alpha=alpha)
plt.plot(worst_s2n_runz['<logAge>_m150'], worst_s2n_runz['<[M/H]>_m150'], 'o', color=pal_wst[2], alpha=alpha)
# plt.xlabel(r"<$\log$ t>", fontsize=fontsize)
plt.ylabel(r"<$[M/H]$>", fontsize=fontsize)
plt.tick_params('both', labelsize=fontsize)
plt.title("mdegree=150", fontsize=fontsize)
plt.xlim(xlim)
plt.ylim(ylim)

ax2 = plt.subplot(1, 3, 2, sharey=ax1)
plt.plot(top_s2n_runz['<logAge>_m200'], top_s2n_runz['<[M/H]>_m200'], 'o', color=pal_top[2], alpha=alpha)
plt.plot(worst_s2n_runz['<logAge>_m200'], worst_s2n_runz['<[M/H]>_m200'], 'o', color=pal_wst[2], alpha=alpha)
plt.xlabel(r"<$\log$ t>", fontsize=fontsize)
plt.tick_params('both', labelsize=fontsize)
plt.title("mdegree=200", fontsize=fontsize)
plt.xlim(xlim)
plt.ylim(ylim)
ax2.yaxis.set_visible(False)

ax3 = plt.subplot(1, 3, 3, sharey=ax1)
plt.plot(top_s2n_runz['<logAge>_m230'], top_s2n_runz['<[M/H]>_m230'], 'o', color=pal_top[2], alpha=alpha)
plt.plot(worst_s2n_runz['<logAge>_m230'], worst_s2n_runz['<[M/H]>_m230'], 'o', color=pal_wst[2], alpha=alpha)
# plt.xlabel(r"<$\log$ t>", fontsize=fontsize)
plt.tick_params('both', labelsize=fontsize)
plt.title("mdegree=230", fontsize=fontsize)
plt.xlim(xlim)
plt.ylim(ylim)
ax3.yaxis.set_visible(False)


plt.savefig(os.path.join(outputs_path, 'Analyses_mdegrees/mdegs_logage_metallicity.png'))

plt.tight_layout()
plt.show()
```
