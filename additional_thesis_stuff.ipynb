# Libraries

```python
import numpy                as np
import pandas               as pd
import matplotlib.pyplot    as plt
import seaborn              as sns
import pyneb                as pn
import os
```

```python
plotpars = {'axes.labelsize': 18,
            'axes.linewidth': 1.,
            'figure.figsize': (10.,8.),
            'font.size': 20,
            'legend.frameon': True,
            'legend.framealpha': 0.85,
            'legend.edgecolor': 'black',
            'legend.loc': 'best',
            'legend.fontsize': 12,
            'mathtext.fontset': 'stixsans',
            'savefig.bbox': 'tight',
            'xtick.labelsize': 18,
            'ytick.labelsize': 18,
           }
```

```python
saving_path = '../LargeFilesResults/'
```

```python
rc = pn.RedCorr()
```

```python
rc.getLaws()
```

```python
rc.printLaws()
```

```python
# %matplotlib notebook
```

```python
plt.rcParams.update(plotpars)
rc.R_V = 3.1
rc.plot(laws=['F99', 'CCM89', 'SM79 Gal'], linewidth=2.)
plt.xlim([1000, 10000])
plt.xlabel(r"Wavelength ($\mathrm{\AA}$)")
plt.ylabel(r"$\zeta$")
plt.tight_layout()
plt.savefig(os.path.join(saving_path, 'extinctionlaws.png'))
plt.savefig(os.path.join(saving_path, 'extinctionlaws.pdf'))
plt.show()
```

```python
plt.rcParams.update(plotpars)
rc.R_V = 3.1
rc.plot(laws=['F99', 'CCM89', 'SM79 Gal'], linewidth=2.)
plt.xlim([1100, 2600])
plt.ylim([0, 12])
plt.xlabel(r"Wavelength ($\mathrm{\AA}$)")
plt.ylabel(r"$\zeta$")
plt.tight_layout()
plt.savefig(os.path.join(saving_path, 'extinctionlaws_zoom.png'))
plt.savefig(os.path.join(saving_path, 'extinctionlaws_zoom.pdf'))
plt.show()
```

```python
rc.R_V = 3.1
plt.rcParams.update(plotpars)

fig, ax = plt.subplots(2, 1)

rc.plot(laws=['F99', 'CCM89', 'SM79 Gal'], linewidth=2., ax=ax[0])
ax[0].set_xlim([1100, 10000])
ax[0].set_xlabel(r"")
ax[0].set_ylabel(r"$\zeta$")

rc.plot(laws=['F99', 'CCM89', 'SM79 Gal'], linewidth=2., ax=ax[1])
plt.xlim([1100, 2400])
plt.ylim([2.5, 10])
plt.xlabel(r"Wavelength ($\mathrm{\AA}$)")
plt.ylabel(r"$\zeta$")

plt.tight_layout()
plt.savefig(os.path.join(saving_path, 'extinctionlaws_all&zoom.png'))
plt.savefig(os.path.join(saving_path, 'extinctionlaws_all&zoom.pdf'))
plt.show()
```
