```python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import seaborn           as sns
import pystan
import arviz
import time
import os
```

```python
data = pd.read_csv('../../Catalogue/Match07_smalldoubleclean_emlines.csv')
```

```python
uv_upturn_temp = data[data['UV_CLASS_YI2011']=='UV_UPTURN']
uv_weak_temp   = data[data['UV_CLASS_YI2011']=='UV_WEAK']
rsf_temp       = data[data['UV_CLASS_YI2011']=='RSF']
uv_upturn_rp   = uv_upturn_temp[uv_upturn_temp['WHAN_CLASS']=='Retired/Passive']
uv_upturn_na   = uv_upturn_temp[uv_upturn_temp['WHAN_CLASS'].isnull()]
uv_upturn      = pd.concat((uv_upturn_na, uv_upturn_rp))
```

```python
# for i in uv_upturn.keys():
#     print(i)
```

```python
colours = {}
colours['CATAID']   = uv_upturn['CATAID'].values
colours['FUV-NUV']  = uv_upturn['MAG_AB_FUV'].values - uv_upturn['MAG_AB_NUV'].values
colours['FUV-R']    = uv_upturn['MAG_AB_FUV'].values - uv_upturn['MAG_AB_R'].values
colours['NUV-R']    = uv_upturn['MAG_AB_NUV'].values - uv_upturn['MAG_AB_R'].values
colours['U-G']      = uv_upturn['MAG_AB_U'].values - uv_upturn['MAG_AB_G'].values
colours['G-R']      = uv_upturn['MAG_AB_G'].values - uv_upturn['MAG_AB_R'].values
colours['R-I']      = uv_upturn['MAG_AB_R'].values - uv_upturn['MAG_AB_I'].values
colours['I-Z']      = uv_upturn['MAG_AB_I'].values - uv_upturn['MAG_AB_Z'].values
colours['redshift'] = uv_upturn['Z']
colours['logmstar'] = uv_upturn['logmstar']
colours = pd.DataFrame(colours)
```

```python
heatmap = colours.drop(['CATAID'], axis=1)
```

```python
cmap = sns.diverging_palette(179, 40, n=10)
# cmap = sns.diverging_palette(150, 50, n=10)
```

```python
plt.rcParams.update(plotpars_hmp)
sns.heatmap(heatmap.corr(method='spearman'), cmap=cmap, vmin=-1, vmax=1, annot=True)
plt.show()
```

```python
plt.rcParams.update(plotpars_1x2)
plt.subplot(1, 2, 1)
plt.scatter(colours['redshift'], colours['G-R'], marker='o')
plt.xlabel(r"redshift")
plt.ylabel(r"$g-r$")

plt.subplot(1, 2, 2)
plt.scatter(colours['logmstar'], colours['G-R'], marker='o')
plt.xlabel(r"$\log M_{\star}$")
plt.ylabel(r"$g-r$")

plt.tight_layout()
plt.show()
```

```python
plt.rcParams.update(plotpars_1x2)
plt.subplot(1, 2, 1)
plt.scatter(data['Z'], data['MAG_AB_G']-data['MAG_AB_R'], marker='o', alpha=0.5)
plt.xlabel(r"redshift")
plt.ylabel(r"$g-r$")

plt.subplot(1, 2, 2)
plt.scatter(data['logmstar'], data['MAG_AB_G']-data['MAG_AB_R'], marker='o', alpha=0.5)
plt.xlabel(r"$\log M_{\star}$")
plt.ylabel(r"$g-r$")
plt.xlim([8,13])

plt.tight_layout()
plt.show()
```

```python
plt.rcParams.update(plotpars_1x2)
plt.subplot(1, 2, 1)
plt.scatter(colours['redshift'], colours['FUV-NUV'], marker='o')
plt.xlabel(r"redshift")
plt.ylabel(r"FUV-NUV")

plt.subplot(1, 2, 2)
plt.scatter(colours['logmstar'], colours['FUV-NUV'], marker='o')
plt.xlabel(r"$\log M_{\star}$")
plt.ylabel(r"FUV-NUV")

plt.tight_layout()
plt.show()
```
