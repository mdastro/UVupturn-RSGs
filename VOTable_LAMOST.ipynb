# Packages

```python
import numpy              as np
import pandas             as pd
import matplotlib.pyplot  as plt
import astropy.io.votable as vot
import astropy.io.fits    as fits
import os
import urllib.request     as r
```

# Match file (GAMA/SDSS/GAMA AND LAMOST)

```python
uv_lamost_dr1 = pd.read_csv('../SPECS_OTHER/SDSS_LAMOST_DR5.csv')
```

```python
uv_lamost_dr1
```

```python
lamost_spectra_url = uv_lamost_dr1['accref']             # field with the address of each spectrum
```

```python
# downloading all files I matched from LAMOST
spec_list = []
for i in lamost_spectra_url:
    name_temp = i.split('/')[-1]
    spec_list.append(name_temp)
    r.urlretrieve(url=i, filename=os.path.join('../SPECS_OTHER/', name_temp))
```

# Opening and checking each file

```python
spec_list = os.listdir('../SPECS_OTHER/')
del(spec_list[5])
```

```python
spec_list
```

```python
for i in spec_list:
    name      = i.split('.')[0]  
    extension = i.split('.')[-1]
    if extension=='vot':
        spec_temp = vot.parse(os.path.join('../SPECS_OTHER/', i))
        spec_temp = spec_temp.get_first_table().array.data
        fx = []
        wl = []
        for j in range(spec_temp.size):
            wl.append(spec_temp[j][0])
            fx.append(spec_temp[j][1])
        spec_temp_df = pd.DataFrame(np.column_stack([np.array(wl), np.array(fx)]))
        spec_temp_df.columns = ['wavelength', 'flux']
        spec_temp_df.to_csv(os.path.join('../SPECS_OTHER/csv', name+'.csv'))
    elif extension=='fits':
        fx = fits.getdata(os.path.join('../SPECS_OTHER/', i))[0,:]
        header_temp = fits.open(os.path.join('../SPECS_OTHER/', i))[0].header
        wl_i    = 10**(header_temp['CRVAL1'])
        wl_step = 10**(header_temp['CD1_1'])
        wl      = np.arange(fx.size)*wl_step+wl_i
        spec_temp_df = pd.DataFrame(np.column_stack([np.array(wl), np.array(fx)]))
        spec_temp_df.columns = ['wavelength', 'flux']
        spec_temp_df.to_csv(os.path.join('../SPECS_OTHER/csv', name+'.csv'), )
    else:
        print("something is really really wrong")
```

# Plotting the spectra
#### Once you have done the above steps, you can skip to the following part of the JN:

```python
csv_files_path = '../SPECS_OTHER/csv'
```

```python
spec_csv_list = os.listdir(csv_files_path)
```

```python
for i in spec_csv_list:
    name = i.split('.')[0]
    csv_temp = pd.read_csv(os.path.join(csv_files_path, i))
    print (csv_temp['wavelength'].values.min(), csv_temp['wavelength'].values.max())
    plt.rcParams.update(plotpars)
    plt.plot(csv_temp['wavelength'], csv_temp['flux'], color='black', label="LAMOST %s" % name)
    plt.legend(loc='best')
    plt.xlabel("Wavelength (\AA)")
    plt.ylabel("Flux")
    plt.savefig(os.path.join('../SPECS_OTHER/figs', name+'.png'))
    plt.show()
```
