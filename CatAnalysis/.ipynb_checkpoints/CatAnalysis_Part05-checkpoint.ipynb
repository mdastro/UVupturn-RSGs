# Packages

```python
import numpy              as np
import pandas             as pd
import matplotlib.pyplot  as plt
import seaborn            as sns
import os
from matplotlib.gridspec  import GridSpec
from mpl_toolkits.mplot3d import Axes3D
from astropy.cosmology    import FlatLambdaCDM
```

# Constants

```python
mpc_to_cm  = 3.086E24
h_alpha_wl = 6.563E-5   # in cm = 6563A
```

# Dataset

```python
data = pd.read_csv('../../Catalogue/Match07_smalldoubleclean_emlines.csv')
```

```python
data = data.fillna('Unclassified')
```

```python
saving_path = '../../Figs/PatriciaSummary/'
```

#### UV upturn objects

```python
uvup_data = data.loc[data['UV_CLASS_YI2011']=='UV_UPTURN']
```

```python
data_gby = uvup_data.groupby('WHAN_CLASS')
```

```python
data_count = data_gby.count()
```

```python
data_count
```

```python
order = [2,0,4,3,1]
```

```python
plt.bar(x=data_count.index.values[order], height=data_count['CATAID'].values[order])
plt.tight_layout()
plt.savefig(os.path.join(saving_path, 'distribution_of_uvups.png'))
plt.show()
```

# Cosmology

```python
adopted_cosmology   = FlatLambdaCDM(H0=70, Om0=0.3)
luminosity_distance = adopted_cosmology.luminosity_distance(uvup_data['Z']).value
```

### Removing spurious H$\alpha$ measurements

```python
rm_spur_halpha = np.where(uvup_data['HA_FLUX_COMP'].values!=-99999.)
```

```python
luminosity = 4*np.pi*((luminosity_distance[rm_spur_halpha]/h_alpha_wl)**2)*(uvup_data['HA_EW_COMP'].
                                                                               iloc[rm_spur_halpha].values)
luminosity_cm = luminosity * mpc_to_cm
```

### SFR

```python
sfr = (7.9E-42) * luminosity_cm # the wl correction 
```

```python
bins = np.arange(start=0, stop=6, step=0.5)
alpha = 0.7
```

```python
plt.hist(sfr[np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='SF')], bins=bins, log=True, alpha=alpha, 
         label='SF')
plt.hist(sfr[np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='wAGN')], bins=bins, log=True, alpha=alpha, 
         label='wAGN')
plt.hist(sfr[np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='sAGN')], bins=bins, log=True, alpha=alpha, 
         label='sAGN')
plt.hist(sfr[np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='Unclassified')], bins=bins, log=True, 
         alpha=alpha, label='Unclassified')
plt.hist(sfr[np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='Retired/Passive')], bins=bins, log=True, 
         alpha=alpha, label='Retired/Passive')
plt.legend(loc='best')
plt.ylabel(r"Frequency")
plt.xlabel(r"SFR")
plt.savefig(os.path.join(saving_path, 'distribution_of_uvups_sfr.png'))
plt.show()
```

```python
idx_sp = np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='Unclassified')
idx_rp = np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='Retired/Passive')
idx_wa = np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='wAGN')
idx_sa = np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='sAGN')
idx_sf = np.where(uvup_data['WHAN_CLASS'].iloc[rm_spur_halpha].values=='SF')

print(len(np.squeeze(idx_sp)))
```

```python
bins2 = bins = np.arange(start=0, stop=0.45, step=0.05)
plt.hist(uvup_data['Z'].iloc[idx_sf], bins=bins2, log=False, alpha=alpha, label='SF')
plt.hist(uvup_data['Z'].iloc[idx_sa], bins=bins2, log=False, alpha=alpha, label='sAGN')
plt.hist(uvup_data['Z'].iloc[idx_wa], bins=bins2, log=False, alpha=alpha, label='wAGN')
plt.hist(uvup_data['Z'].iloc[idx_rp], bins=bins2, log=False, alpha=alpha, label='Retired/Passive')
plt.hist(uvup_data['Z'].iloc[idx_sp], bins=bins2, log=False, alpha=alpha, label='Unclassified')

plt.legend(loc='best')
plt.ylabel(r"Frequency")
plt.xlabel(r"Redshift")
plt.savefig(os.path.join(saving_path, 'distribution_of_uvups_redshift.png'))
plt.show()
```

```python
plt.semilogy(uvup_data['logmstar'].iloc[idx_sf], sfr[idx_sf], 'o', label='SF', alpha=0.5)
plt.semilogy(uvup_data['logmstar'].iloc[idx_sa], sfr[idx_sa], 'o', label='sAGN', alpha=0.5)
plt.semilogy(uvup_data['logmstar'].iloc[idx_wa], sfr[idx_wa], 'o', label='wAGN', alpha=0.5)
plt.semilogy(uvup_data['logmstar'].iloc[idx_rp], sfr[idx_rp], 'o', label='Retired/Passive', alpha=0.5)
plt.semilogy(uvup_data['logmstar'].iloc[idx_sp], sfr[idx_sp], 'o', label='Unclassified', alpha=0.5)
plt.title(r"UV upturn (by Yi et al. 2011) and their WHAN class")

plt.xlabel('Log(M$_*$)')
plt.ylabel('SFR (M$_\odot$ yr$^{-1}$)')
plt.legend(loc='best')
plt.savefig(os.path.join(saving_path, 'mass_sfr_uvup.png'))
plt.show()

# idea: clustering test?
```

```python
plt.semilogy(uvup_data['logage'].iloc[idx_sf], sfr[idx_sf], 'o', label='SF', alpha=0.5)
plt.semilogy(uvup_data['logage'].iloc[idx_sa], sfr[idx_sa], 'o', label='sAGN', alpha=0.5)
plt.semilogy(uvup_data['logage'].iloc[idx_wa], sfr[idx_wa], 'o', label='wAGN', alpha=0.5)
plt.semilogy(uvup_data['logage'].iloc[idx_rp], sfr[idx_rp], 'o', label='Retired/Passive', alpha=0.5)
plt.semilogy(uvup_data['logage'].iloc[idx_sp], sfr[idx_sp], 'o', label='Unclassified', alpha=0.5)
plt.title(r"UV upturn (by Yi et al. 2011) and their WHAN class")

plt.xlabel('Log(t)')
plt.ylabel('SFR (M$_\odot$ yr$^{-1}$)')
plt.legend(loc='upper right')
plt.xlim([9.7, 10.])
plt.savefig(os.path.join(saving_path, 'age_sfr_uvup.png'))
plt.show()

# idea: clustering test?
```
