```python
data_up = data[data['UV_CLASS_YI2011']=='UV_UPTURN']
data_wk = data[data['UV_CLASS_YI2011']=='UV_WEAK']
```

```python
n, bins, patches = plt.hist(data_wk['SurfaceDensity'], bins=20, label="Weak", color='yellow', alpha=0.9)
plt.hist(data_up['SurfaceDensity'], bins=bins, label='Upturn', color='red', alpha=0.5)
plt.yscale('log')
plt.show()
```

```python
sns.kdeplot(data_wk['SurfaceDensity'], cumulative=True, bw='scott')
sns.kdeplot(data_up['SurfaceDensity'], cumulative=True, bw='scott')
plt.xlim([-10, 40])
plt.show()
```
